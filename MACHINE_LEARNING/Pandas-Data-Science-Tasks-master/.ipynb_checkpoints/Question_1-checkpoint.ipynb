```python
all_data = pd.read_csv('all_datas.csv')
all_data.head(3)
```

## steps to solve this problem
1. split the months and create a new col
2. add new col for the total sale of every product
3. the use groupby to get the sum of every month

# find the NAN and drop it

```python
# get the NaN cell
nan_cells = all_data[all_data['Month'].isna()]
nan_cells

all_data = all_data.dropna(how='all')
all_data.head()
```

# Clean the 'Or' at the month col
### Do not Drop the whole row just use != 'Or'

```python
all_data['Month'] = all_data['Order Date'].str[0:2]

all_data = all_data[all_data['Month'] != 'Or']

all_data['Month'] = all_data['Month'].astype(int)
```

```python
all_data.head()
```

### Add salse col by multipling the [Quantity Ordered with Price Each]

### first convert the Quantity Ordered to int

```python
all_data['Salse'] = all_data['Quantity Ordered'].astype(int) * all_data['Price Each']
all_data
```

### now use the groupby to get the filtered sum of every month

```python
sumSalesOf_eachMonth = all_data.groupby('Month').sum()
```
