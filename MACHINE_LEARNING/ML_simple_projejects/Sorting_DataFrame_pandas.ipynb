```python
import pandas as pd

# import thoes two files 
df = pd.read_csv('data/survey_results_public.csv')
schecma_df = pd.read_csv('data/survey_results_schema.csv')
```

# Sorting

```python
# sorting by index
df.sort_index(ascending=False)


# for sorting by a specific column
df.sort_values(by='ConvertedComp', ascending=False)


# sorting with multiple columns =====> we can use inplace=True
df.sort_values(by=['Country', 'ConvertedComp'], ascending=[False, False]).head(4)


# sorting the 7 smallest salaries in the survey
df['ConvertedComp'].nsmallest(7)


# sorting the 7 highest salaries in the survey
df['ConvertedComp'].nlargest(7)


# OR we can like below if u want to have his survey as well.
df.nlargest(3, 'ConvertedComp')


# insase u can NOT use the inplace=True JUST assigne it to the old df
```
